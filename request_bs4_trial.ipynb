## BeautifulSoup 을 통한 태그 탐색

### select rule
1. .class
1. #id
1. tag

```python
from bs4 import BeautifulSoup as soup
```

```python
news = soup(html, "html.parser")
```

```python
news_list = news.select_one(".newsList") 
news_head = news_list.select(".articleSubject > a")
news_press = news_list.select(".articleSummary > .press")
news_date = news_list.select(".articleSummary > .wdate")
# news_press.select("#press")
```

```python
news_title = []
for i in range(len(news_head)):
    news_title.append([news_head[i].get_text()
                      ,news_press[i].get_text()
                      ,news_date[i].get_text()])
    
df = pd.DataFrame(news_title, columns=["title", "press", "date"])
```

## 30일치의 뉴스

```python
days=[]
for i in range(30):
    
    date = datetime.now() - timedelta(days=int("{0}".format(i)))
    date = date.strftime("%Y-%m-%d")
    days.append(date)
days[:5]
```

```python
news_title = []

for day in days:
    url = f"https://finance.naver.com/news/mainnews.nhn?date={day}"
    response = requests.get(url)
    html = response.text
    news = soup(html, "html.parser")
    
    news_list = news.select_one(".newsList") 
    news_head = news_list.select(".articleSubject > a")
    news_press = news_list.select(".articleSummary > .press")
    news_date = news_list.select(".articleSummary > .wdate")
    
    

    for i in range(len(news_head)):
        news_title.append([news_head[i].get_text()
                          ,news_press[i].get_text()
                          ,news_date[i].get_text()])
        
    random_sleep = np.random.uniform(0,2) # 0~2초 사이에 쉬면서 가져온다
    time.sleep(random_sleep)
    
news_title
```
